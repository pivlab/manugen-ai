# Demonstrate agentic output

The goal of this notebook is to help demonstrate agentic output from this project.
We can use this output to observe through review that the work is acting as we think it
should.

```python
import pathlib
import tempfile

from IPython.display import Image, display
from manugen_ai.agents.repo_to_paper.agent import root_agent
from manugen_ai.utils import build_mermaid, run_agent_workflow
```

```python
# show our loop as a mermaid diagram
_, PNG = build_mermaid(root_agent)
display(Image(PNG))
```

```python
# generate some example data
temp_dir = tempfile.TemporaryDirectory()
dir_path = pathlib.Path(temp_dir.name)
# Create example markdown files
(dir_path / "README.md").write_text("# Example Project\n\nThis is a sample README.")
(dir_path / "intro.md").write_text("# Introduction\n\nWelcome to the example.")
(dir_path / "methods.md").write_text("# Methods\n\nDetails about the methods.")
dir_path
```

```python
APP_NAME = "db_tool_app"
USER_ID = "repo_to_paper_user"
SESSION_ID = "session_0001"
target_repo = "https://github.com/cytomining/CytoTable"
final_output, session_state, output_events = await run_agent_workflow(
    agent=root_agent,
    prompt=f"""
    Please turn this repository and a filepath provided in the sate 
    into a scientific paper abstract and revise it within a loop.
    Here is the filepath: {dir_path}

    Here is the repository:
    {target_repo}
    """,
    app_name=APP_NAME,
    user_id=USER_ID,
    session_id=SESSION_ID,
    initial_state={
        "repository_url": target_repo,
        "content_path": temp_dir,
    },
    # note: use `uv run pytest -s` to view the verbose stdout
    verbose=True,
)
```
